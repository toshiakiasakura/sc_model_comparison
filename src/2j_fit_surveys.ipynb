## Estimate EVIs

```julia
include("main_utils.jl")
include("fit_utils.jl")
```

```julia
df_dds = CSV.read("../dt_surveys_master/master_dds.csv", DataFrame);
```

```julia
keys_ = df_dds[:, :key] |> unique
qs = [0.95, 0.97, 0.98]

Random.seed!(1234)
res_hm = collect_estimates_across_studies(df_dds, keys_; strat = "home", qs=qs)
res_hm[:, :strat] .= "home"
Random.seed!(1234)
res_nhm = collect_estimates_across_studies(df_dds, keys_; strat = "non-home", qs=qs)
res_nhm[:, :strat] .= "non-home"

res = vcat(res_hm, res_nhm)
CSV.write("../dt_intermediate/EVI_summary.csv", res)
```
